```python
# Import library yang dibutuhkan
import os
import numpy as np
import librosa
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Input, GlobalAveragePooling2D
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt

# Atur seed untuk reproduksibilitas awal
np.random.seed(42)
tf.random.set_seed(42)

# Sampling rate standar
SR = 22050
NUM_RUNS = 30 # Training masing-masing model 30 kali

# Fungsi untuk padding audio agar memiliki panjang yang sama
def pad_audio(y, target_length):
    """
    Padding dengan nol agar audio memiliki panjang `target_length`.
    """
    if len(y) < target_length:
        y = np.pad(y, (0, target_length - len(y)), mode='constant')
    else:
        y = y[:target_length]
    return y

import numpy as np

# def pad_audio_center(y, target_length):
#     """
#     Memusatkan audio di tengah array target_length,
#     dengan padding nol di kiri dan kanan.
#     """
#     if len(y) >= target_length:
#         # Potong jika terlalu panjang
#         start = (len(y) - target_length) // 2
#         y = y[start : start + target_length]
#     else:
#         # Hitung padding kiri dan kanan
#         total_padding = target_length - len(y)
#         pad_left = total_padding // 2
#         pad_right = total_padding - pad_left
#         y = np.pad(y, (pad_left, pad_right), mode='constant')
#     return y


# 1. Load Dataset
def load_audio_files(dataset_path, sr=SR, duration=None):
    """
    Memuat file audio dan label dari folder dataset.
    Asumsi: folder `dataset_path` memiliki subfolder untuk setiap kelas,
    masing-masing berisi file .wav.
    """
    audio_data = []
    labels = []
    class_names = []
    # Ambil daftar kelas (nama folder)
    for class_idx, class_name in enumerate(sorted(os.listdir(dataset_path))):
        class_folder = os.path.join(dataset_path, class_name)
        if not os.path.isdir(class_folder):
            continue
        class_names.append(class_name)
        # Iterasi file .wav dalam folder kelas
        for file_name in os.listdir(class_folder):
            if file_name.endswith('.wav'):
                file_path = os.path.join(class_folder, file_name)
                # Muat audio menggunakan librosa
                y, _ = librosa.load(file_path, sr=sr, duration=duration)
                audio_data.append(y)
                labels.append(class_idx)
    return audio_data, np.array(labels), class_names

# 2. Ekstraksi Fitur
def extract_mel_spectrogram(audio, sr=SR, n_mels=128, fmax=8000):
    """
    Menghitung mel-spectrogram log dari sinyal audio.
    """
    S = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels, fmax=fmax)
    S_db = librosa.power_to_db(S, ref=np.max)
    return S_db

def extract_mfcc(audio, sr=SR, n_mfcc=40):
    """
    Menghitung koefisien MFCC dari sinyal audio.
    """
    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    return mfcc

# 3. Membangun Model CNN untuk mel-spectrogram
def create_cnn_model_mel(input_shape, num_classes):
    """
    Membangun model CNN sederhana untuk klasifikasi berdasar mel-spectrogram.
    """
    model = Sequential([
        Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2,2)),
        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(128, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    return model

# 4. Membangun Model CNN untuk MFCC
def create_cnn_model_mfcc(input_shape, num_classes):
    """
    Membangun model CNN sederhana untuk klasifikasi berdasar MFCC.
    """
    model = Sequential([
        Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2,2)),
        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Conv2D(128, (3,3), activation='relu'),
        MaxPooling2D((2,2)),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    return model

# Fungsi plotting
def plot_training_history(history, title="Model", run=None):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(1, len(acc)+1)

    plt.figure(figsize=(12,5))
    run_title = f" (Run {run})" if run else ""

    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title(f'{title} - Accuracy{run_title}')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title(f'{title} - Loss{run_title}')

    plt.tight_layout()
    plt.show()

def resize_feature(feature, target_shape=(224, 224)):
    """
    Resize/padding fitur agar berukuran target_shape (224x224).
    """
    current_shape = feature.shape
    pad_height = max(0, target_shape[0] - current_shape[0])
    pad_width = max(0, target_shape[1] - current_shape[1])

    # Padding ke target shape
    padded = np.pad(feature,
                      pad_width=((0, pad_height), (0, pad_width)),
                      mode='constant', constant_values=0)

    # Jika lebih besar, crop
    resized = padded[:target_shape[0], :target_shape[1]]

    return resized

# ------------------------------------------------------------------------
# Main pipeline
# ------------------------------------------------------------------------
if __name__ == '__main__':
    # Path ke dataset
    dataset_path = 'Trimmed Audio Dataset'

    # 1) Muat data audio dan label 
    print("Memuat data audio...")
    audio_data, labels, class_names = load_audio_files(dataset_path, sr=SR)
    num_classes = len(class_names)

    # 1.a) Padding setiap audio ke panjang maksimum
    max_len = max(len(y) for y in audio_data)
    audio_data = [pad_audio(y, max_len) for y in audio_data]
    audio_data = np.array(audio_data)

    # Pra-pemrosesan fitur 
    print("Mengekstrak fitur Mel-Spectrogram dan MFCC...")
    X_mel_full = []
    for y in audio_data:
        S_db = extract_mel_spectrogram(y, sr=SR, n_mels=128)
        resized_feature = resize_feature(S_db, target_shape=(224, 224))
        X_mel_full.append(resized_feature[..., np.newaxis])
    X_mel_full = np.array(X_mel_full)

    X_mfcc_full = []
    for y in audio_data:
        mfcc = extract_mfcc(y, sr=SR, n_mfcc=40)
        resized_feature = resize_feature(mfcc, target_shape=(224, 224))
        X_mfcc_full.append(resized_feature[..., np.newaxis])
    X_mfcc_full = np.array(X_mfcc_full)

    # Simpan metrik untuk semua model dari semua run
    all_run_results = []
    all_run_cms = []

    # Definisikan callbacks
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=10, # sedikit lebih sabar
        restore_best_weights=True,
        verbose=1
    )
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )
    callbacks = [early_stop, reduce_lr]

    EPOCHS = 150
    BATCH_SIZE = 16

    # Loop untuk melatih dan mengevaluasi model sebanyak NUM_RUNS
    for run in range(1, NUM_RUNS + 1):
        print(f"\n{'='*30} RUN {run}/{NUM_RUNS} {'='*30}")

        # 1.b) Tentukan pembagian train-test dengan stratifikasi label untuk setiap run
        # Menggunakan random_state berbeda agar split bervariasi tapi tetap reproduktif
        indices = np.arange(len(labels))
        train_idx, test_idx = train_test_split(indices, test_size=0.2, stratify=labels, random_state=42 + run)
        
        y_train, y_test = labels[train_idx], labels[test_idx]

        # ========================================================================
        # Model 1: CNN dengan fitur mel-spectrogram
        # ========================================================================
        print(f"\n[RUN {run} - Model 1] CNN dengan fitur Mel-Spectrogram")
        
        # Bagi train/test
        X_train_mel, X_test_mel = X_mel_full[train_idx], X_mel_full[test_idx]
        
        # Bangun, compile, dan latih model (harus di dalam loop untuk reset weights)
        input_shape_mel = X_train_mel.shape[1:]
        model_mel = create_cnn_model_mel(input_shape_mel, num_classes)
        model_mel.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        
        print(f"Training Mel-CNN model for run {run}...")
        history_mel = model_mel.fit(
            X_train_mel, y_train,
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_test_mel, y_test),
            callbacks=callbacks,
            verbose=0 # Set ke 0 agar tidak terlalu ramai, atau 2 untuk ringkasan per epoch
        )
        
        # Tampilkan grafik untuk run ini
        plot_training_history(history_mel, title="CNN Mel-Spectrogram", run=run)
        
        # Evaluasi
        print(f"Evaluating Mel-CNN model for run {run}...")
        y_pred_mel = np.argmax(model_mel.predict(X_test_mel), axis=1)
        
        # Kumpulkan metrik performa
        run_results_mel = {
            "Run": run,
            "Model": "CNN Mel",
            "Accuracy": accuracy_score(y_test, y_pred_mel),
            "Precision": precision_score(y_test, y_pred_mel, average='macro', zero_division=0),
            "Recall": recall_score(y_test, y_pred_mel, average='macro', zero_division=0),
            "F1": f1_score(y_test, y_pred_mel, average='macro', zero_division=0),
            "Train Accuracy": history_mel.history['accuracy'][-1],
            "Train Loss": history_mel.history['loss'][-1],
            "Val Accuracy": history_mel.history['val_accuracy'][-1],
            "Val Loss": history_mel.history['val_loss'][-1]
        }
        all_run_results.append(run_results_mel)
        
        # Kumpulkan confusion matrix
        cm_mel = confusion_matrix(y_test, y_pred_mel)
        for true_idx, row in enumerate(cm_mel):
            for pred_idx, count in enumerate(row):
                all_run_cms.append({
                    'Run': run,
                    'Model': 'CNN Mel',
                    'True_Label': class_names[true_idx],
                    'Predicted_Label': class_names[pred_idx],
                    'Count': count
                })

        print(f"Classification Report (Mel-CNN, Run {run}):")
        print(classification_report(y_test, y_pred_mel, target_names=class_names, zero_division=0))

        # ========================================================================
        # Model 2: CNN dengan fitur MFCC
        # ========================================================================
        print(f"\n[RUN {run} - Model 2] CNN dengan fitur MFCC")
        X_train_mfcc, X_test_mfcc = X_mfcc_full[train_idx], X_mfcc_full[test_idx]
        
        model_mfcc = create_cnn_model_mfcc(X_train_mfcc.shape[1:], num_classes)
        model_mfcc.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        
        print(f"Training MFCC-CNN model for run {run}...")
        history_mfcc = model_mfcc.fit(
            X_train_mfcc, y_train,
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_test_mfcc, y_test),
            callbacks=callbacks,
            verbose=0
        )
        
        # Tampilkan grafik untuk run ini
        plot_training_history(history_mfcc, title="CNN MFCC", run=run)
        
        # Evaluasi
        print(f"Evaluating MFCC-CNN model for run {run}...")
        y_pred_mfcc = np.argmax(model_mfcc.predict(X_test_mfcc), axis=1)

        # Kumpulkan metrik performa
        run_results_mfcc = {
            "Run": run,
            "Model": "CNN MFCC",
            "Accuracy": accuracy_score(y_test, y_pred_mfcc),
            "Precision": precision_score(y_test, y_pred_mfcc, average='macro', zero_division=0),
            "Recall": recall_score(y_test, y_pred_mfcc, average='macro', zero_division=0),
            "F1": f1_score(y_test, y_pred_mfcc, average='macro', zero_division=0),
            "Train Accuracy": history_mfcc.history['accuracy'][-1],
            "Train Loss": history_mfcc.history['loss'][-1],
            "Val Accuracy": history_mfcc.history['val_accuracy'][-1],
            "Val Loss": history_mfcc.history['val_loss'][-1]
        }
        all_run_results.append(run_results_mfcc)
        
        # Kumpulkan confusion matrix
        cm_mfcc = confusion_matrix(y_test, y_pred_mfcc)
        for true_idx, row in enumerate(cm_mfcc):
            for pred_idx, count in enumerate(row):
                all_run_cms.append({
                    'Run': run,
                    'Model': 'CNN MFCC',
                    'True_Label': class_names[true_idx],
                    'Predicted_Label': class_names[pred_idx],
                    'Count': count
                })
        
        print(f"Classification Report (MFCC-CNN, Run {run}):")
        print(classification_report(y_test, y_pred_mfcc, target_names=class_names, zero_division=0))


    # ========================================================================
    # Selesai semua run, simpan hasil kompilasi
    # ========================================================================
    print("\n\n" + "="*25 + " SEMUA RUN SELESAI " + "="*25)

    # Simpan performa ke CSV
    performance_df = pd.DataFrame(all_run_results)
    performance_csv_path = 'all_runs_performance.csv'
    performance_df.to_csv(performance_csv_path, index=False)
    print(f"\nPerforma dari {NUM_RUNS} run disimpan ke: {performance_csv_path}")
    print("Ringkasan Performa (rata-rata dari semua run):")
    print(performance_df.groupby('Model')[['Accuracy', 'Precision', 'Recall', 'F1']].mean())

    # Simpan confusion matrix semua run ke file CSV terpisah
    cm_df = pd.DataFrame(all_run_cms)
    cm_csv_path = 'all_runs_confusion_matrices.csv'
    cm_df.to_csv(cm_csv_path, index=False)
    print(f"\nConfusion matrices dari semua run disimpan ke: {cm_csv_path}")
    
    # ========================================================================
    # Simpan model dari RUN TERAKHIR
    # ========================================================================
    
    # Pastikan folder 'fix_Model/CNN_Kombinasi1' ada
    model_save_path = 'fix_Model/CNN_Kombinasi1'
    os.makedirs(model_save_path, exist_ok=True)
    
    # Simpan dalam format native Keras (.keras)
    model_mel.save(os.path.join(model_save_path, 'Mel1_final_run.keras'))
    model_mfcc.save(os.path.join(model_save_path, 'MFCC1_final_run.keras'))
    print(f"\nModel dari run terakhir ({NUM_RUNS}) disimpan di folder: {model_save_path}")
```
