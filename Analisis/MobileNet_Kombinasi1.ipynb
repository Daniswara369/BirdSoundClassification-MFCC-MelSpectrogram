```python
# Import library yang dibutuhkan
import os
import numpy as np
import librosa
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, GlobalAveragePooling2D
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import pandas as pd
import matplotlib.pyplot as plt

# Atur seed untuk reproduksibilitas awal
np.random.seed(42)
tf.random.set_seed(42)

# Konfigurasi
SR = 22050
NUM_RUNS = 30 # Training masing-masing model 30 kali

# --- Fungsi Helper (Sama seperti sebelumnya) ---

def pad_audio(y, target_length):
    if len(y) < target_length:
        y = np.pad(y, (0, target_length - len(y)), mode='constant')
    else:
        y = y[:target_length]
    return y

def load_audio_files(dataset_path, sr=SR):
    audio_data = []
    labels = []
    class_names = []
    for class_idx, class_name in enumerate(sorted(os.listdir(dataset_path))):
        class_folder = os.path.join(dataset_path, class_name)
        if not os.path.isdir(class_folder):
            continue
        class_names.append(class_name)
        for file_name in os.listdir(class_folder):
            if file_name.endswith('.wav'):
                file_path = os.path.join(class_folder, file_name)
                y, _ = librosa.load(file_path, sr=sr)
                audio_data.append(y)
                labels.append(class_idx)
    return audio_data, np.array(labels), class_names

def extract_mel_spectrogram(audio, sr=SR, n_mels=128, fmax=8000):
    S = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels, fmax=fmax)
    S_db = librosa.power_to_db(S, ref=np.max)
    return S_db

def extract_mfcc(audio, sr=SR, n_mfcc=40):
    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    return mfcc

def resize_feature(feature, target_shape=(224, 224)):
    current_shape = feature.shape
    # Untuk resize, kita bisa menggunakan tf.image.resize yang lebih efisien
    # Tambahkan dimensi channel sementara
    feature_with_channel = feature[..., np.newaxis]
    resized = tf.image.resize(feature_with_channel, target_shape).numpy()
    return np.squeeze(resized, axis=-1)

def plot_training_history(history, title="Model", run=None):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(1, len(acc)+1)

    plt.figure(figsize=(12, 5))
    run_title = f" (Run {run})" if run else ""
    
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title(f'{title} - Accuracy{run_title}')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title(f'{title} - Loss{run_title}')

    plt.tight_layout()
    plt.show()

# --- Fungsi Pembangunan Model ---

def create_transfer_mobilenet(input_shape, num_classes):
    """
    Membangun model klasifikasi dengan MobileNet (pra-terlatih) dari Keras Applications.
    """
    # Menggunakan `weights='imagenet'` secara otomatis akan mengunduh bobot yang sesuai.
    base_model = tf.keras.applications.MobileNet(input_shape=input_shape, include_top=False, weights='imagenet')
    base_model.trainable = False
    
    # Membangun model baru di atas base_model
    inputs = Input(shape=input_shape)
    # MobileNet mengharapkan input ternormalisasi. Kita bisa menggunakan lapisan normalisasi
    # atau membiarkannya belajar dari data float yang ada. Untuk konsistensi, kita biarkan saja.
    x = base_model(inputs, training=False)
    x = GlobalAveragePooling2D()(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs, outputs)
    return model

# ------------------------------------------------------------------------
# Main pipeline
# ------------------------------------------------------------------------
if __name__ == '__main__':
    dataset_path = 'Trimmed Audio Dataset'
    
    # 1) Muat dan pra-pemrosesan data (hanya sekali)
    print("Memuat data audio...")
    audio_data, labels, class_names = load_audio_files(dataset_path, sr=SR)
    num_classes = len(class_names)
    
    max_len = max(len(y) for y in audio_data)
    audio_data = [pad_audio(y, max_len) for y in audio_data]

    # Ekstraksi fitur dan persiapan untuk model (hanya sekali)
    print("Mengekstrak fitur dan mempersiapkannya untuk MobileNet...")
    X_mel_full = []
    for y in audio_data:
        S_db = extract_mel_spectrogram(y, sr=SR)
        resized_feature = resize_feature(S_db, target_shape=(224, 224))
        # Ulangi channel untuk membuat input 3-channel
        rgb_feature = np.repeat(resized_feature[..., np.newaxis], 3, axis=-1)
        X_mel_full.append(rgb_feature)
    X_mel_full = np.array(X_mel_full)

    X_mfcc_full = []
    for y in audio_data:
        mfcc = extract_mfcc(y, sr=SR)
        resized_feature = resize_feature(mfcc, target_shape=(224, 224))
        rgb_feature = np.repeat(resized_feature[..., np.newaxis], 3, axis=-1)
        X_mfcc_full.append(rgb_feature)
    X_mfcc_full = np.array(X_mfcc_full)
    
    print(f"Bentuk data Mel-Spectrogram: {X_mel_full.shape}")
    print(f"Bentuk data MFCC: {X_mfcc_full.shape}")

    # Inisialisasi list untuk menyimpan hasil dari semua run
    all_run_results = []
    all_run_cms = []

    # Definisikan callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6, verbose=1)
    callbacks = [early_stop, reduce_lr]

    EPOCHS = 120
    BATCH_SIZE = 16

    # Loop untuk melatih dan mengevaluasi model sebanyak NUM_RUNS
    for run in range(1, NUM_RUNS + 1):
        print(f"\n{'='*30} RUN {run}/{NUM_RUNS} {'='*30}")

        indices = np.arange(len(labels))
        train_idx, test_idx = train_test_split(indices, test_size=0.2, stratify=labels, random_state=42 + run)
        y_train, y_test = labels[train_idx], labels[test_idx]
        
        # --- Model 1: MobileNet dengan Mel-Spectrogram ---
        print(f"\n[RUN {run} - Model 1] Transfer Learning MobileNet (Mel-spectrogram)")
        X_train_mel, X_test_mel = X_mel_full[train_idx], X_mel_full[test_idx]
        
        model_melnet = create_transfer_mobilenet(X_train_mel.shape[1:], num_classes)
        model_melnet.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

        history_melnet = model_melnet.fit(
            X_train_mel, y_train,
            epochs=EPOCHS, batch_size=BATCH_SIZE, validation_data=(X_test_mel, y_test),
            callbacks=callbacks, verbose=0
        )
        
        plot_training_history(history_melnet, title="MobileNet Mel-Spectrogram", run=run)
        
        y_pred_melnet = np.argmax(model_melnet.predict(X_test_mel), axis=1)
        
        # Kumpulkan Metrik Performa
        all_run_results.append({
            "Run": run, "Model": "MobileNet (Mel)",
            "Accuracy": accuracy_score(y_test, y_pred_melnet),
            "Precision": precision_score(y_test, y_pred_melnet, average='macro', zero_division=0),
            "Recall": recall_score(y_test, y_pred_melnet, average='macro', zero_division=0),
            "F1": f1_score(y_test, y_pred_melnet, average='macro', zero_division=0),
            "Train Accuracy": history_melnet.history['accuracy'][-1], "Train Loss": history_melnet.history['loss'][-1],
            "Val Accuracy": history_melnet.history['val_accuracy'][-1], "Val Loss": history_melnet.history['val_loss'][-1]
        })
        
        # Kumpulkan Confusion Matrix
        cm_melnet = confusion_matrix(y_test, y_pred_melnet)
        for true_idx, row in enumerate(cm_melnet):
            for pred_idx, count in enumerate(row):
                all_run_cms.append({'Run': run, 'Model': 'MobileNet (Mel)', 'True_Label': class_names[true_idx], 'Predicted_Label': class_names[pred_idx], 'Count': count})
        
        print(f"Classification Report (MobileNet-Mel, Run {run}):\n", classification_report(y_test, y_pred_melnet, target_names=class_names, zero_division=0))

        # --- Model 2: MobileNet dengan MFCC ---
        print(f"\n[RUN {run} - Model 2] Transfer Learning MobileNet (MFCC)")
        X_train_mfcc, X_test_mfcc = X_mfcc_full[train_idx], X_mfcc_full[test_idx]
        
        model_mfccnet = create_transfer_mobilenet(X_train_mfcc.shape[1:], num_classes)
        model_mfccnet.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        
        history_mfccnet = model_mfccnet.fit(
            X_train_mfcc, y_train,
            epochs=EPOCHS, batch_size=BATCH_SIZE, validation_data=(X_test_mfcc, y_test),
            callbacks=callbacks, verbose=0
        )

        plot_training_history(history_mfccnet, title="MobileNet MFCC", run=run)

        y_pred_mfccnet = np.argmax(model_mfccnet.predict(X_test_mfcc), axis=1)
        
        # Kumpulkan Metrik Performa
        all_run_results.append({
            "Run": run, "Model": "MobileNet (MFCC)",
            "Accuracy": accuracy_score(y_test, y_pred_mfccnet),
            "Precision": precision_score(y_test, y_pred_mfccnet, average='macro', zero_division=0),
            "Recall": recall_score(y_test, y_pred_mfccnet, average='macro', zero_division=0),
            "F1": f1_score(y_test, y_pred_mfccnet, average='macro', zero_division=0),
            "Train Accuracy": history_mfccnet.history['accuracy'][-1], "Train Loss": history_mfccnet.history['loss'][-1],
            "Val Accuracy": history_mfccnet.history['val_accuracy'][-1], "Val Loss": history_mfccnet.history['val_loss'][-1]
        })

        # Kumpulkan Confusion Matrix
        cm_mfccnet = confusion_matrix(y_test, y_pred_mfccnet)
        for true_idx, row in enumerate(cm_mfccnet):
            for pred_idx, count in enumerate(row):
                all_run_cms.append({'Run': run, 'Model': 'MobileNet (MFCC)', 'True_Label': class_names[true_idx], 'Predicted_Label': class_names[pred_idx], 'Count': count})

        print(f"Classification Report (MobileNet-MFCC, Run {run}):\n", classification_report(y_test, y_pred_mfccnet, target_names=class_names, zero_division=0))

    # ========================================================================
    # Selesai semua run, simpan hasil kompilasi
    # ========================================================================
    print("\n\n" + "="*25 + " SEMUA RUN SELESAI " + "="*25)

    # Simpan performa ke CSV
    performance_df = pd.DataFrame(all_run_results)
    performance_csv_path = 'mobilenet_kombinasi1_all_performance.csv'
    performance_df.to_csv(performance_csv_path, index=False)
    print(f"\nPerforma dari {NUM_RUNS} run disimpan ke: {performance_csv_path}")
    print("Ringkasan Performa (rata-rata dari semua run):")
    print(performance_df.groupby('Model')[['Accuracy', 'Precision', 'Recall', 'F1']].mean())

    # Simpan confusion matrix semua run ke file CSV terpisah
    cm_df = pd.DataFrame(all_run_cms)
    cm_csv_path = 'mobilenet_kombinasi1_all_confusion_matrices.csv'
    cm_df.to_csv(cm_csv_path, index=False)
    print(f"\nConfusion matrices dari semua run disimpan ke: {cm_csv_path}")
    
    # ========================================================================
    # Simpan model dari RUN TERAKHIR
    # ========================================================================
    model_save_path = 'fix_Model/MobileNet_Kombinasi1'
    os.makedirs(model_save_path, exist_ok=True)
    
    model_melnet.save(os.path.join(model_save_path, 'MobileNet_Mel_final_run.keras'))
    model_mfccnet.save(os.path.join(model_save_path, 'MobileNet_MFCC_final_run.keras'))
    print(f"\nModel dari run terakhir ({NUM_RUNS}) disimpan di folder: {model_save_path}")

```
