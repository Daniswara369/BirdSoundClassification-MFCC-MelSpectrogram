```python
# Import library yang dibutuhkan
import os
import numpy as np
import librosa
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers, models
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import pandas as pd
import matplotlib.pyplot as plt

# Atur seed untuk reproduksibilitas awal
np.random.seed(42)
tf.random.set_seed(42)

# Konfigurasi
SR = 22050
NUM_RUNS = 30 # 

# --- Fungsi Helper ---

def pad_audio(y, target_length):
    """Padding dengan nol agar audio memiliki panjang `target_length`."""
    if len(y) < target_length:
        y = np.pad(y, (0, target_length - len(y)), mode='constant')
    else:
        y = y[:target_length]
    return y

def load_audio_files(dataset_path, sr=SR):
    """Memuat file audio dan label dari folder dataset."""
    audio_data = []
    labels = []
    class_names = []
    for class_idx, class_name in enumerate(sorted(os.listdir(dataset_path))):
        class_folder = os.path.join(dataset_path, class_name)
        if not os.path.isdir(class_folder):
            continue
        class_names.append(class_name)
        for file_name in os.listdir(class_folder):
            if file_name.endswith('.wav'):
                file_path = os.path.join(class_folder, file_name)
                y, _ = librosa.load(file_path, sr=sr)
                audio_data.append(y)
                labels.append(class_idx)
    return audio_data, np.array(labels), class_names

def plot_training_history(history, title="Model", run=None):
    """Membuat plot akurasi dan loss dari riwayat training."""
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(1, len(acc)+1)

    plt.figure(figsize=(12, 5))
    run_title = f" (Run {run})" if run else ""
    
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title(f'{title} - Accuracy{run_title}')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title(f'{title} - Loss{run_title}')

    plt.tight_layout()
    plt.show()

# --- Fungsi Spesifik untuk VGGish ---

def extract_vggish_embeddings(audio_list, sr=SR, target_sr=16000):
    """
    Menggunakan VGGish dari TensorFlow Hub untuk ekstraksi embedding.
    Proses ini memakan waktu dan hanya perlu dijalankan sekali.
    """
    print("Memuat model VGGish dari TensorFlow Hub...")
    vggish = hub.load('https://tfhub.dev/google/vggish/1')
    print("Model VGGish berhasil dimuat.")

    embeddings = []
    print(f"Mengekstrak VGGish embeddings dari {len(audio_list)} file audio...")
    for i, y in enumerate(audio_list):
        # Resample ke 16 kHz yang dibutuhkan oleh VGGish
        if sr != target_sr:
            y_res = librosa.resample(y, orig_sr=sr, target_sr=target_sr)
        else:
            y_res = y.astype(np.float32)

        # Dapatkan embeddings dari VGGish
        # VGGish menghasilkan embedding per 0.96 detik, jadi kita ambil rata-ratanya
        emb_frames = vggish(y_res)
        mean_emb = tf.reduce_mean(emb_frames, axis=0).numpy()
        embeddings.append(mean_emb)
        if (i + 1) % 10 == 0:
            print(f"  ... Selesai {i + 1}/{len(audio_list)}")

    print("Ekstraksi embedding selesai.")
    return np.stack(embeddings, axis=0)

def create_classifier_from_embeddings(input_dim, num_classes):
    """
    Membangun MLP sederhana untuk klasifikasi berdasarkan vektor embedding.
    """
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    return model

# ------------------------------------------------------------------------
# Main pipeline
# ------------------------------------------------------------------------
if __name__ == '__main__':
    dataset_path = 'Trimmed Audio Dataset'
    
    # 1) Muat data audio dan label (hanya sekali)
    print("Memuat data audio...")
    audio_data, labels, class_names = load_audio_files(dataset_path, sr=SR)
    num_classes = len(class_names)
    
    max_len = max(len(y) for y in audio_data)
    audio_data = [pad_audio(y, max_len) for y in audio_data]

    # 2) Ekstraksi fitur embedding VGGish (hanya sekali, karena prosesnya berat)
    X_vggish_full = extract_vggish_embeddings(audio_data, sr=SR)
    print(f"Bentuk data embedding VGGish setelah ekstraksi: {X_vggish_full.shape}")

    # 
    all_run_results = []
    all_run_cms = []

    # Definisikan callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6, verbose=1)
    callbacks = [early_stop, reduce_lr]

    EPOCHS = 200
    BATCH_SIZE = 16

    # Loop untuk melatih dan mengevaluasi model klasifikasi sebanyak NUM_RUNS
    for run in range(1, NUM_RUNS + 1):
        print(f"\n{'='*30} RUN {run}/{NUM_RUNS} {'='*30}")

        # Bagi data embedding yang sudah diekstrak
        X_train_vgg, X_test_vgg, y_train, y_test = train_test_split(
            X_vggish_full, labels, test_size=0.2, stratify=labels, random_state=42 + run
        )
        
        # Bangun, compile, dan latih model klasifikasi (di-reset setiap run)
        model_vgg = create_classifier_from_embeddings(input_dim=X_train_vgg.shape[1], num_classes=num_classes)
        model_vgg.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        
        print(f"Training VGGish classifier for run {run}...")
        history_vgg = model_vgg.fit(
            X_train_vgg, y_train,
            epochs=EPOCHS, batch_size=BATCH_SIZE, validation_data=(X_test_vgg, y_test),
            callbacks=callbacks, verbose=0
        )
        
        # Tampilkan grafik untuk run ini
        plot_training_history(history_vgg, title="VGGish Classifier", run=run)
        
        # Evaluasi model
        print(f"Evaluating VGGish classifier for run {run}...")
        y_pred_vgg = np.argmax(model_vgg.predict(X_test_vgg), axis=1)
        
        # Kumpulkan metrik performa
        all_run_results.append({
            "Run": run, "Model": "VGGish",
            "Accuracy": accuracy_score(y_test, y_pred_vgg),
            "Precision": precision_score(y_test, y_pred_vgg, average='macro', zero_division=0),
            "Recall": recall_score(y_test, y_pred_vgg, average='macro', zero_division=0),
            "F1": f1_score(y_test, y_pred_vgg, average='macro', zero_division=0),
            "Train Accuracy": history_vgg.history['accuracy'][-1], "Train Loss": history_vgg.history['loss'][-1],
            "Val Accuracy": history_vgg.history['val_accuracy'][-1], "Val Loss": history_vgg.history['val_loss'][-1]
        })
        
        # Kumpulkan confusion matrix
        cm_vgg = confusion_matrix(y_test, y_pred_vgg)
        for true_idx, row in enumerate(cm_vgg):
            for pred_idx, count in enumerate(row):
                all_run_cms.append({'Run': run, 'Model': 'VGGish', 'True_Label': class_names[true_idx], 'Predicted_Label': class_names[pred_idx], 'Count': count})
        
        print(f"Classification Report (VGGish, Run {run}):\n", classification_report(y_test, y_pred_vgg, target_names=class_names, zero_division=0))

    # ========================================================================
    # Selesai semua run, simpan hasil kompilasi
    # ========================================================================
    print("\n\n" + "="*25 + " SEMUA RUN SELESAI " + "="*25)

    # Simpan performa ke CSV
    performance_df = pd.DataFrame(all_run_results)
    performance_csv_path = 'vggish1_all_run_performance.csv'
    performance_df.to_csv(performance_csv_path, index=False)
    print(f"\nPerforma VGGish dari {NUM_RUNS} run disimpan ke: {performance_csv_path}")
    print("Ringkasan Performa VGGish (rata-rata dari semua run):")
    print(performance_df[['Accuracy', 'Precision', 'Recall', 'F1']].mean())

    # Simpan confusion matrix semua run ke file CSV terpisah
    cm_df = pd.DataFrame(all_run_cms)
    cm_csv_path = 'vggish1_all_run_confusion_matrices.csv'
    cm_df.to_csv(cm_csv_path, index=False)
    print(f"\nConfusion matrices VGGish dari semua run disimpan ke: {cm_csv_path}")
    
    # ========================================================================
    # Simpan model klasifikasi dari RUN TERAKHIR
    # ========================================================================
    model_save_path = 'fix_Model/VGGish'
    os.makedirs(model_save_path, exist_ok=True)
    
    # Simpan model klasifikasi (bukan model VGGish-nya)
    model_vgg.save(os.path.join(model_save_path, 'VGGish1_final_run.keras'))
    print(f"\nModel klasifikasi VGGish dari run terakhir ({NUM_RUNS}) disimpan di folder: {model_save_path}")
```
